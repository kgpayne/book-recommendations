```python
import pickle
import pandas as pd
import recordlinkage

from pprint import pprint
from book_tools.index import Soundex
```

```python
bot = pd.read_csv('01_books_of_titans_clean.csv', index_col=[0])
bot = bot.fillna('')
bot.head()
```

```python
aud = pd.read_csv('01_audiobooks_clean.csv', index_col=[0])
aud = aud.fillna('')
aud.head()
```

```python
# Index by Soundex of title and author_1_surname
indexer = recordlinkage.Index()
indexer.add(Soundex('title'))
indexer.add(Soundex('author_1_surname'))
candidate_pairs = indexer.index(bot, aud)
candidate_pairs
```

```python
# Write candidate pairs
with open('02_candidate_pairs.p', 'wb') as outfile:
    pickle.dump(candidate_pairs, outfile)
```

```python
# Calculate Comparison Reduction
```

```python
# Full comparison
from scipy.special import comb
full_comp = int(comb(len(bot) + len(aud), 2))
full_comp
```

```python
# Reduced comparison
red_comp = len(candidate_pairs)
red_comp
```

```python
# % Comparison Reduction
100 - ((red_comp/full_comp)*100)
```
