```python
import pandas as pd
```

```python
bot = pd.read_csv('01_cleaned_books_of_titans.csv', index_col=[0])
bot = bot.fillna('')
bot.head(1)
```

```python
clusters = pd.read_csv('04_match_clusters.csv', index_col=[0])
clusters.head(1)
```

```python
df = (
    bot.merge(clusters, how='left', left_index=True, right_index=True)
    .sort_values(by=['cluster_size', 'cluster_id', 'title'], ascending=[False, True, True])
)
df.head(127)
```
