```python
import json
import pandas as pd
```

```python
with open('theindependentbookseller_awards_database.json', 'r') as awards_db:
    response = json.load(awards_db)
```

```python
response.keys()
```

```python
response['data']
```

```python
data = response['data']
```

```python
genres = next(col for col in data['columns'] if col['name'] == 'Genre/Category')['typeOptions']['choices']
genre_map = {choice['id']:choice['name'] for _, choice in genres.items()}
genre_map
```

```python
len(data['columns'])
```

```python
data['columns'][:2]
```

```python
len(data['rows'])
```

```python
data['rows'][:5]
```

```python
extracted_rows = [
    row['cellValuesByColumnId'] for row in data['rows']
]
extracted_rows[:1]
```

```python
name_map = {col['id']:col['name'] for col in data['columns']}
name_map
```

```python
cleaned_rows = [
    {name_map[k]: (v if not name_map[k] == 'Genre/Category' else genre_map[v]) for k, v in row.items()} for row in extracted_rows
]
cleaned_rows[:5]
```

```python
df = pd.DataFrame(cleaned_rows)
df.head()
```
