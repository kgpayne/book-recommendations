```python
# Merge Tim Ferris list with Derek Sivers reviews
```

```python
df = pd.merge(tf, ds, on='base_title', how='outer')
```

```python
df['recommendations'] = df['podcast_no'].apply(lambda x: len(x) if isinstance(x, list) else 0)
```

```python
df.sort_values(['recommendations', 'rating'], ascending=[False, False])
```

```python
mtchs = df[~pd.isna(df['rating']) & (df['recommendations'] != 0)]
len(mtchs)
```
