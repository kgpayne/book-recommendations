```python
transl_table = dict( [ (ord(x), ord(y)) for x,y in zip( u"‘’´“”–-",  u"'''\"\"--") ] ) 
ppl['people_mentioned_name'] = ppl['people_mentioned_name'].str.translate(transl_table)
```

```python
ppl[ppl['people_mentioned_name'].str.contains('"')]
```

```python
def quotes_to_bracktes(text):
    new_text = ''
    first_bracket = True
    for char in text:
        if char == '"' and first_bracket:
            new_text += '('
            first_bracket = False
        elif char == '"' and not first_bracket:
            new_text += ')'
            first_bracket = False
        else:
            new_text += char
    return new_text
```

```python
ppl['people_mentioned_name'] = ppl['people_mentioned_name'].apply(lambda x: quotes_to_bracktes(x))
```
